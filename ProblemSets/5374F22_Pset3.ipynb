1. Let $A$ and $B$ be two conjugate $n\times n$ matrices, i.e. $A = M^{-1}BM$ for some invertible square matrix $M$. Show that 
    \begin{align*}
      e^{A} = M^{-1}e^{B}M
    \end{align*}
    Use this to show that if $A$ is a symmetric matrix then $\det(e^{A})=e^{\text{tr}(A)}$ for all $t$, and in particular, if we further assume that $\text{tr}(A)=0$, then the linear transformation $e^{tA}$ is a measure preserving transformation for each $t$.
    
2. Let $A$ be a $m\times n$ matrix and $b \in \mathbb{R}^n$. Given any $x_0 \in \mathbb{R}^n$ and a unit vector $e\in \mathbb{R}^n$, define the function
    \begin{align*}
      f_{x,e}(t) = \|A(x_0+te)-b\|_2^2
    \end{align*}
    Show that $f_{x,e}$ is a convex quadratic polynomial in $t$, and write down the coefficients of this polynomial in terms of $A,b,x_0$, and $e$. Use this to find the $t$ producing the minimum value for $f_{x,e}$ in terms of these same quantities.
